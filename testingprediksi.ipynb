```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
```

```python
df_train = pd.read_csv('data training.csv', encoding='utf-8')
df_test = pd.read_csv('data testing.csv', encoding='utf-8')
```

```python
vectorizer = TfidfVectorizer()
```

```python
tfidf_train = vectorizer.fit_transform(df_train['Isi Surat'].astype(str))
tfidf_test = vectorizer.transform(df_test['Isi Surat'].astype(str))
```

```python
from scipy.sparse import hstack

X_train = hstack([df_train[['Selisih Hari']].values, tfidf_train])
X_test = hstack([df_test[['Selisih Hari']].values, tfidf_test])

y_train = df_train['Jenis Surat']
```

```python
knn = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
df_test['Predicted Jenis Surat'] = y_pred
```

```python
print("=== Hasil Prediksi Jenis Surat ===")
print(df_test[['Isi Surat', 'Selisih Hari', 'Predicted Jenis Surat']])
```
